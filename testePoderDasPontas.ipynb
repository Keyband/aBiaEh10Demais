```python
#Importar bibliotecas
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Primeiro plot: superfície aproximadamente oval
epsilon=0.1
m=2
k=100

thetaArray = np.linspace(0,2*math.pi,k)
rArray = 1 + epsilon*np.cos(m*thetaArray)
sigmaArray = 1 + epsilon*(m-1)*np.cos(m*thetaArray)
KArray = 1 + epsilon*(m**2 - 1)*np.cos(m*thetaArray)

fig = plt.figure(figsize=(5, 10))

ax = plt.subplot(2, 1, 1)
for theta, r, sigma, K in zip(thetaArray, rArray, sigmaArray, KArray):
    plt.plot(theta, r, 'ro')
    plt.plot(theta, sigma, 'bx')
    plt.plot(theta, K, 'g-')

ax = plt.subplot(2, 1, 2, projection='polar')
for theta, r, sigma, K in zip(thetaArray, rArray, sigmaArray, KArray):
    plt.polar(theta, r, 'ro')
    plt.polar(theta, sigma, 'bx')
    plt.polar(theta, K, 'gs')

plt.show()
```

```python
#Terceiro plot: Tentativa de plot interativo
#Começamos definindo umas funções
from ipywidgets import interactive

plotSize=3.5
k=100
thetaArray = np.linspace(0,2*math.pi,k)

def rFunction(m,epsilon,theta):
    return 1 + epsilon*np.cos(m*theta)

def sigmaFunction(m,epsilon,theta):
    return 1 + epsilon*(m-1)*np.cos(m*theta)

def KFunction(m,epsilon,theta):
    return 1 + epsilon*(m**2 - 1)*np.cos(m*theta)

def VFunction(m,epsilon,theta):
    r=rFunction(m,epsilon,theta)
    return np.log(r)+epsilon*(1/r**m)*np.cos(m*theta)

def polarRPlot(eps=epsilon,n=m):
    ax=plt.subplot(111,projection='polar')
    ax.plot(thetaArray, rFunction(n,eps,thetaArray))
    ax.set_frame_on(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim(ymax=plotSize)
    
def polarSigmaPlot(eps=epsilon,n=m):
    ax=plt.subplot(111,projection='polar')
    ax.plot(thetaArray, sigmaFunction(n,eps,thetaArray))
    ax.set_frame_on(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim(ymax=plotSize)
    
def polarKPlot(eps=epsilon,n=m):
    ax=plt.subplot(111,projection='polar')
    ax.plot(thetaArray, KFunction(n,eps,thetaArray))
    ax.set_frame_on(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim(ymax=plotSize)
    
def polarVPlot(eps=epsilon,n=m):
    ax=plt.subplot(111,projection='polar')
    ax.plot(thetaArray, VFunction(n,eps,thetaArray))
    ax.set_frame_on(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim(ymax=plotSize)
```

```python
#Plot de r
interactive(polarRPlot,eps=(0.05,0.5,0.025),n=(1,10))
```

```python
#Plot de sigma
interactive(polarSigmaPlot,eps=(0.05,0.5,0.025),n=(1,10))
```

```python
#Plot de K
interactive(polarKPlot,eps=(0.05,0.5,0.025),n=(1,10))
```

```python
#Plot de V
interactive(polarVPlot,eps=(0.05,0.5,0.025),n=(1,10))
```
